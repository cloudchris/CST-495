# Term Frequency

We will first start with smaller test set of the movie descriptions

```python
corpus = items_d[0:20]
print corpus
```

Compute the frequency of the entire corpus

```python
tf = {}
for doc in corpus:
        for word in doc.split():
            if word in tf:
                tf[word] += 1
            else:
                tf[word] = 1
print tf
```

Below will compute the same as above but will use the Counter collection

```python
from collections import Counter

def get_tf(corpus):
    tf = Counter()
    for doc in corpus:
        for word in doc.split():
            tf[word] += 1
    return tf

tf = get_tf(corpus)
print(tf)
```

Enter the entire movie description into get_tf to determing frequency of 'action', 'adventure', 'documentary'

```python
tf = get_tf(items_d)
print tf['action']
print tf['adventure']
print tf['documentary']
```

# Document Term Frequency

Compute term frequency for each document.

```python
import collections

def get_tf(document):
    tf = Counter()
    for word in document.split():
        tf[word] += 1
    return tf

def get_dtf(corpus):
    dtf = {}
    for i, doc in enumerate(corpus):
        dtf[i]=get_tf(doc)
    return dtf

dtf = get_dtf(items_d)
dtf[87]
```
